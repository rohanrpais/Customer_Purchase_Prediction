```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

# PREPROCESSING

```python
path=r'D:\social_network_ads\data\dataset.xlsx'
df=pd.read_excel(path)
```

```python
df.drop(columns=['Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6'])
```

```python
x=df[['Age','Estimated Salary']]
y=df['Purchased']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=36)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# MODELS 

```python
from sklearn.metrics import (
    accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
)
import time 
```

```python
# Log Reg
from sklearn.linear_model import LogisticRegression
start_time=time.time()
log_reg=LogisticRegression(
    max_iter=100,
    random_state=36
)
log_reg.fit(x_train,y_train)
train_time=time.time()-start_time
print("training time is ",train_time)
```

```python
start_time=time.time()
y_predict=log_reg.predict(x_test)
predict_time=time.time()-start_time
print("prediction time is ",predict_time)
```

```python
accuracy=accuracy_score(y_test,y_predict)
precision=precision_score(y_test,y_predict)
recall=recall_score(y_test,y_predict)
f1=f1_score(y_test,y_predict)
con_mat=confusion_matrix(y_test,y_predict)
print(
    "accuracy: ",accuracy,
    "\nprecision: ",precision,
    "\nrecall: ",recall,
    "\nf1 score: ",f1,
    "\nconfusion matrix : ",con_mat
)
```

```python
results=[]
results.append({
    "Model":"Logistic Regression",
    "Train Time":train_time,
    "Predict Time":predict_time,
    "Accuracy": accuracy,
    "Precision":precision,
    "Recall":recall,
    "F1 score":f1,
})
```
