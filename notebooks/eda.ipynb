# EDA of the Dataset

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
path=r'D:\social_network_ads\data\dataset.xlsx'
df=pd.read_excel(path)
```

```python
df=df.drop(columns=['Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6',])
```

```python
print(df.columns)
print(df['Purchased'].value_counts())
print(df['Estimated Salary'].value_counts())
print(df['Age'].value_counts())
```

```python
df2 =df.groupby('Purchased')['Estimated Salary'].value_counts()
print(df2)
df3=df.groupby('Purchased')['Age'].value_counts()
print(df3)
```

```python
print(df.isnull().sum())
print((df['Age']<0).sum())
print((df['Estimated Salary']<0).sum())
print((df['Purchased']<0).sum())
```
